```python
import numpy as np

# 4 tokens, embedding size = 3
X = np.array([
    [1.0, 0.0, 1.0],   # I
    [1.0, 1.0, 0.0],   # love
    [0.0, 1.0, 1.0],   # deep
    [0.0, 0.0, 1.0],   # learning
])
```

```python
print(X)
```

```python
X.T
```

```python
# For similrity dot product
scores = X @ X.T
print(scores)
```

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return exp_x / exp_x.sum(axis=-1, keepdims=True)

weights = softmax(scores)
print(weights)
```
