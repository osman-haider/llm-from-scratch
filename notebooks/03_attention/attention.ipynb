# Evolution of Sequence Modeling in NLP

Early neural machine translation and sequence modeling systems were based on **RNN encoderdecoder** architectures. In this setup, the encoder processes the input sequence and compresses it into a fixed-length context vector, which is then passed to the decoder to generate the output sequence.

However, this approach suffers from a context bottleneck, especially for long sequences, because all input information must be stored in a single vector. As sequence length increases, important information can be lost, leading to degraded performance.

To address this limitation, **Bahdanau et al**. (2014) introduced the attention mechanism. Instead of relying solely on a fixed context vector, the decoder dynamically computes a weighted combination of all encoder hidden states at each decoding step. This allows the model to focus on the most relevant parts of the input sequence, significantly improving performance on long sequences.

While attention improved RNN-based models, recurrence still imposed limitations on parallelization and long-range dependency modeling. To overcome these issues, **Transformers** (Vaswani et al., 2017) were introduced. Transformers eliminate recurrence entirely and rely on self-attention mechanisms to model dependencies between all tokens in a sequence simultaneously. This enables efficient parallel computation and more effective modeling of long-range relationships.

**Self-attention** is a mechanism that allows each position in the input sequence to attend to all positions in the same sequence when computing the representation of a sequence. Self-attention is a key component of contemporary LLMs based on the transformer architecture, such as the GPT series.

As a result, Transformer-based architectures have become the foundation of most modern NLP system.
